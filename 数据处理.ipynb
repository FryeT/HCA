```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import glob
```

```python
files = glob.glob('datas\*.csv')
df = pd.read_csv(files[1])

t = 'Time since start of test (s)'
u = 'Pore Pressure (kPa)'
tau_ztheta = 'HCA Tau Theta Z'
gamma = 'HCA Gamma'
epsilon_1 = 'HCA E1'
epsilon_2 = 'HCA E2'
epsilon_3 = 'HCA E3'
font1 = {'family': 'Times New Roman', 'size': 16}
```

```python
fig1 = plt.figure(figsize=(9,3))
ax = fig1.add_subplot(1,1,1)
ax.plot(0.1*df[t], df[u]-df[u][0], 'k')

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(which='both', direction='in', labelsize=12)
# ax.minorticks_on()

ax.set_xlim(left=0)
ax.set_ylim(bottom=0)
ax.set_xlabel('N', fontdict=font1)
ax.set_ylabel('u / kPa', fontdict=font1)
ax.xaxis.set_label_coords(1.02,0.06)
```

```python
fig2 = plt.figure(figsize=(9,3))
ax = fig2.add_subplot(1,1,1)
ax.plot(0.1*df[t], df[tau_ztheta], 'k')

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(which='both', direction='in', labelsize=12)
# ax.minorticks_on()
ax.set_xlim(left=0)
ax.set_xlabel('N', fontdict=font1)
ax.set_ylabel(r'$\tau_{z\theta}$ / kPa', fontdict=font1)
ax.xaxis.set_label_coords(1.02,0.06)
```

```python
fig4 = plt.figure(figsize=(9,3))
ax = fig4.add_subplot(1,1,1)
ax.plot(df[gamma], df[tau_ztheta], 'k')

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(which='both', direction='in', labelsize=12)
# ax.minorticks_on()
ax.set_xlabel(r'$\gamma_{z\theta}$ /%', fontdict=font1)
ax.set_ylabel(r'$\tau_{z\theta}$ / kPa', fontdict=font1)
```

```python
fig5 = plt.figure(figsize=(9,3))
ax = fig5.add_subplot(1,1,1)
e1 = df[epsilon_1] - df[epsilon_1][0]
e2 = df[epsilon_2] - df[epsilon_2][0]
e3 = df[epsilon_3] - df[epsilon_3][0]
gamma_g = np.sqrt(2)/3*np.sqrt((e1-e2)**2 + (e1-e3)**2 + (e2-e3)**2)
ax.plot(0.1*df[t], gamma_g, 'K')

ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(which='both', direction='in', labelsize=12)
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(integer=True))
# ax.minorticks_on()
ax.set_xlim(left=0)
ax.set_ylim(bottom=0)
ax.set_xlabel('N', fontdict=font1)
ax.set_ylabel(r'$\gamma_g$ /%', fontsize = 16)
ax.xaxis.set_label_coords(1.02,0.06)
```

```python
def plot_tg(f, ax, start=0, end=50):
    df = pd.read_csv(f)
    ax.plot(df[gamma][start:end], df[tau_ztheta][start:end], 'k')
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.spines['bottom'].set_position(('data',0))
    ax.spines['left'].set_position(('data',0))
    ax.tick_params(which='both', direction='in', labelsize=12)
    ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(min_n_ticks=4))
    ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(min_n_ticks=5, step=[2], nbins=4))

    ax.set_xlabel(r'$\gamma_{z\theta}$ /%', fontsize=16)
    ax.set_ylabel(r'$\tau_{z\theta}$ / kPa', fontsize=16, rotation='horizontal')
    ax.xaxis.set_label_coords(1.15,0.5)
    ax.yaxis.set_label_coords(0.5,1.02)
```

```python
f = f_0[0]
fig, ax = plt.subplots(1, 1, figsize=(5,3))
plot_tg(f, ax, end=150)
fig.savefig('figs\\' + f[6:-4] + '.png', bbox_inches='tight')
```

```python
from scipy import stats
```

```python
df_m = pd.read_csv('datas/modulus.csv')
df_m
```

```python
g_m = df_m['']
G_d = df_m['']
d_r = df_m['']
e_m = df_m['']
E_d = df_m['']
```

```python
x = g_m[:-1]
y = 1/G_d[:-1]
linetext = 'a={0:.3f},  b={1:.2f}\n$R^2$={2:.3f}'
fig, ax = plt.subplots(1, 1, figsize=(6,3.5))
b, a, r, p, std_err = stats.linregress(x, y)
y_lr = b*x + a
ax.plot(x*1e4, y, 'sk')
ax.plot(x*1e4, y_lr, 'k')
ax.text(7, 0.015, linetext.format(a,b,r*r), fontsize=12)
ax.set_xlim(left=0)   
ax.set_ylim(bottom=0.01)
ax.tick_params(which='both', direction='in', labelsize=12)
ax.set_xlabel(r'$\gamma_m$ /$10^{-4}$', fontsize=16)
ax.set_ylabel(r'$1/G_d$ /$MPa^{-1}$', fontsize=16)
fig.savefig('figs/modulus.png', bbox_inches='tight')
```

```python
x = g_m[:-1]/100
y = 1/G_d[:-1]
linetext = 'a={0:.2e},  b={1:.2e}\n$R^2$={2:.3f}'
fig, ax = plt.subplots(1, 1, figsize=(6,3.5))
b, a, r, p, std_err = stats.linregress(x, y)
y_lr = b*x + a
ax.plot(x, y, 'sk')
ax.plot(x, y_lr, 'k')
ax.text(7, 0.015, linetext.format(a,b,r*r), fontsize=12)
ax.set_xlim(left=0)   
ax.set_ylim(bottom=0.01)
ax.tick_params(which='both', direction='in', labelsize=12)
ax.set_xlabel(r'$\gamma_m$ /$10^{-4}$', fontsize=16)
ax.set_ylabel(r'$1/G_d$ /$MPa^{-1}$', fontsize=16)
fig.savefig('figs/modulus2.png')
```
